```python

# å°å…¥å¿…è¦çš„å¥—ä»¶
import paho.mqtt.client as mqtt
import time
import json
```

```python
# MQTT Broker è¨­å®šï¼ˆé€£æ¥åˆ°æœ¬åœ° Raspberry Pi çš„ MQTT Brokerï¼‰
BROKER = "localhost"  # æˆ–ä½¿ç”¨ "127.0.0.1"
PORT = 1883  # æ¨™æº– MQTT åŸ ï¼Œå¦‚æœæ‚¨çš„ Broker ä½¿ç”¨ä¸åŒåŸ è™Ÿè«‹ä¿®æ”¹
TOPIC = "room/12"  # è¨‚é–±çš„ä¸»é¡Œåç¨±ï¼ˆèˆ‡ publish ç¨‹å¼ç›¸åŒï¼‰

# å»ºç«‹ MQTT å®¢æˆ¶ç«¯ï¼ˆä½¿ç”¨æ–°çš„å›èª¿ API ç‰ˆæœ¬ 2ï¼‰
client = mqtt.Client(callback_api_version=mqtt.CallbackAPIVersion.VERSION2)

# é€£ç·šå›èª¿å‡½æ•¸ï¼ˆVERSION2 APIï¼‰
def on_connect(client, userdata, flags, reason_code, properties):
    if reason_code.is_failure:
        print(f"âŒ é€£ç·šå¤±æ•—ï¼ŒéŒ¯èª¤ä»£ç¢¼: {reason_code}")
    else:
        print(f"âœ… æˆåŠŸé€£æ¥åˆ° MQTT Broker: {BROKER}")
        # é€£ç·šæˆåŠŸå¾Œè¨‚é–±ä¸»é¡Œ
        client.subscribe(TOPIC, qos=1)
        print(f"âœ… å·²è¨‚é–±ä¸»é¡Œ: {TOPIC}")

# è¨Šæ¯æ¥æ”¶å›èª¿å‡½æ•¸ï¼ˆVERSION2 APIï¼‰
def on_message(client, userdata, message):
    topic = message.topic
    payload = message.payload.decode('utf-8')
    qos = message.qos
    
    print(f"\nğŸ“¨ æ”¶åˆ°è¨Šæ¯:")
    print(f"   ä¸»é¡Œ: {topic}")
    print(f"   QoS: {qos}")
    print(f"   å…§å®¹: {payload}")
    
    # å˜—è©¦è§£æ JSON æ ¼å¼
    try:
        data = json.loads(payload)
        print(f"   JSON è§£ææˆåŠŸ: {data}")
    except json.JSONDecodeError:
        print(f"   (ç´”æ–‡å­—è¨Šæ¯)")

# è¨‚é–±å›èª¿å‡½æ•¸ï¼ˆVERSION2 APIï¼‰
def on_subscribe(client, userdata, mid, reason_codes, properties):
    print(f"âœ… è¨‚é–±ç¢ºèªï¼Œè¨Šæ¯ ID: {mid}")

# è¨­å®šå›èª¿å‡½æ•¸
client.on_connect = on_connect
client.on_message = on_message
client.on_subscribe = on_subscribe

# é€£æ¥åˆ° Broker
print(f"æ­£åœ¨é€£æ¥åˆ° {BROKER}...")
client.connect(BROKER, PORT, 60)
client.loop_start()  # é–‹å§‹èƒŒæ™¯åŸ·è¡Œç·’è™•ç†è¨Šæ¯

# ç­‰å¾…é€£ç·šå»ºç«‹
time.sleep(1)
æ­£åœ¨é€£æ¥åˆ° localhost...
æ­£åœ¨é€£æ¥åˆ° localhost...
âœ… æˆåŠŸé€£æ¥åˆ° MQTT Broker: localhost
âœ… å·²è¨‚é–±ä¸»é¡Œ: room/12
âœ… è¨‚é–±ç¢ºèªï¼Œè¨Šæ¯ ID: 1
```
