# MQTT Subscribe 測試應用程式

這個 notebook 示範如何使用 paho-mqtt 套件來訂閱並接收 MQTT 訊息。


```python
# 導入必要的套件
import paho.mqtt.client as mqtt
import time
import json
```

```python
# MQTT Broker 設定（連接到本地 Raspberry Pi 的 MQTT Broker）
BROKER = "localhost"  # 或使用 "127.0.0.1"
PORT = 1883  # 標準 MQTT 埠，如果您的 Broker 使用不同埠號請修改
TOPIC = "客廳/溫度"  # 訂閱的主題名稱（與 publish 程式相同）

# 建立 MQTT 客戶端（使用新的回調 API 版本 2）
client = mqtt.Client(callback_api_version=mqtt.CallbackAPIVersion.VERSION2)

# 連線回調函數（VERSION2 API）
def on_connect(client, userdata, flags, reason_code, properties):
    if reason_code.is_failure:
        print(f"❌ 連線失敗，錯誤代碼: {reason_code}")
    else:
        print(f"✅ 成功連接到 MQTT Broker: {BROKER}")
        # 連線成功後訂閱主題
        client.subscribe(TOPIC, qos=1)
        print(f"✅ 已訂閱主題: {TOPIC}")

# 訊息接收回調函數（VERSION2 API）
def on_message(client, userdata, message):
    topic = message.topic
    payload = message.payload.decode('utf-8')
    qos = message.qos
    
    print(f"\n📨 收到訊息:")
    print(f"   主題: {topic}")
    print(f"   QoS: {qos}")
    print(f"   內容: {payload}")
    
    # 嘗試解析 JSON 格式
    try:
        data = json.loads(payload)
        print(f"   JSON 解析成功: {data}")
    except json.JSONDecodeError:
        print(f"   (純文字訊息)")

# 訂閱回調函數（VERSION2 API）
def on_subscribe(client, userdata, mid, reason_codes, properties):
    print(f"✅ 訂閱確認，訊息 ID: {mid}")

# 設定回調函數
client.on_connect = on_connect
client.on_message = on_message
client.on_subscribe = on_subscribe

# 連接到 Broker
print(f"正在連接到 {BROKER}...")
client.connect(BROKER, PORT, 60)
client.loop_start()  # 開始背景執行緒處理訊息

# 等待連線建立
time.sleep(1)
```

```python
# 持續監聽訊息（測試用）
# 執行此 cell 後，程式會持續監聽訊息
# 可以執行 lesson6_1.ipynb 的 publish 程式來發送測試訊息

print("🔍 開始監聽訊息...")
print("   請執行 lesson6_1.ipynb 的 publish 程式來發送測試訊息")
print("   或使用其他 MQTT 客戶端發送訊息到主題: " + TOPIC)
print("   按 Ctrl+C 或中斷執行來停止監聽\n")

try:
    # 持續監聽 30 秒（可自行調整時間）
    time.sleep(180)
except KeyboardInterrupt:
    print("\n⏹️  停止監聽")
finally:
    print("✅ 監聽結束")
```

```python
# 取消訂閱（可選）
# 如果只想取消訂閱但保持連線，可以使用：
# client.unsubscribe(TOPIC)

# 關閉連線
client.loop_stop()
client.disconnect()
print("✅ MQTT 連線已關閉")
```
